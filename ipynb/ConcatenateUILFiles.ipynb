```python
import numpy as np
import collections
```

```python
filenamesStem=["LX_C_","LX_CH_"]
targetFileStem="LX_CC_"
```

```python
def list_duplicates(seq):
    tally = collections.defaultdict(list)
    for i,item in enumerate(seq):
        tally[item].append(i)
    return ((key,locs) for key,locs in tally.items() 
                            if len(locs)>1)
```

```python
def dupliLines(arr):
    L=list(arr.transpose()[0])
    dupList=[dup for dup in sorted(list_duplicates(L))]
    LL=[dup[1][1] for dup in dupList]
    return LL
```

```python
def removedups(array, linesList):
    return np.delete(array,linesList,axis=0)
```

```python
def removeduplicates(array):
    return removedups(array,dupliLines(array))
```

```python
for m in range(1,5):
    filenames=[stem+str(m) for stem in filenamesStem]
    arrayList=[fileToNpArray(filename, separator="\t", skiplines=1, fileEnding=".uil")[0] for filename in filenames]
    dupFreeArrayList=[removeduplicates(array) for array in arrayList]
    array=np.concatenate(dupFreeArrayList,axis=0)
    array2=removeduplicates(array)
    npArrayToFile(targetFileStem+str(m),array2,preString="Voltage [V]\tCurrent [A]\tPhotocurrent [A]\n",separator="\t", fileEnding=".uil")
```
