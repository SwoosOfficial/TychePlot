```python
import sys
import os
sys.path.append("../Calculator/")
from Filereader import fileToNpArray
import matplotlib.pyplot as plt
```

```python
filenamestem="XLV_B_"
```

```python
sepList=["\t",",",";"]
for m in range(1,5):
    for sep in sepList:
        for n in range(0,200):
            try:
                data=fileToNpArray(filenamestem+str(m), separator=sep, skiplines=n)[0]
                break
            except Exception as err :
                #print(err)
                pass
    for x in range(1,len(data[0])):
        x=int(x)
        plt.plot(data[:,0],data[:,x], label=str(m))
plt.yscale("log")
plt.legend()
plt.show()
```
