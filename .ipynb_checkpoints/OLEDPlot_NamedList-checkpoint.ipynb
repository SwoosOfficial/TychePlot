```python
import sys
sys.path.append("../Calculator/")
from PDF import PDF
from OLEDPlot import OLEDPlot
from OLEDPlot import OLEDNamedListPlot
from MultiPlot import MultiPlot #Multiprocessing
```

```python
#obligatory
spectraPath="/home/tassilo/Dokumente/data/20171204_XXXVIII_JETI_Irppy/XXXVIII_C_100muA.csv"
prefix="XXXVIII"
title="Ir(ppy)$_\\text{3}$"
name=prefix
labelList=["Ir(ppy)$_\\text{3}$ OLED "]
```

```python
#optional
customLims=False # better pick False running first Time
xLims=[[],[-2,10],[-2,6],[-2,6],[2.3,6],[2.3,6],[2.3,6]]
yLims=[[],[-2,10],[10**-7,1],[10**-2,10**5],[0,250],[0,600],[0,25]]
alphaOffset=3
samples=1
colorOffset=1
showErrorOnlyEvery=10
specBg=None
show=[[True,True]]*samples
errors=[[True,True]]*samples
customFontsize=[12,12,8,10,10]
#scaled
scaleX=0.47
scaledCustomFontsize=[10,10,4,6,6]
ax2Labels=False
titleBool=False
```

```python
def initPlot(n, m=0):
    if customLims:
        plot=OLEDNamedListPlot(prefix,
                               labelList,
                               name, 
                               xLim=xLims[n],
                               showColAxLim=yLims, 
                               showColTup=(n,m), 
                               spectraFile=spectraPath, 
                               show=show, 
                               errors=errors, 
                               customFontsize=customFontsize, 
                               alphaOffset=alphaOffset, 
                               samples=samples, 
                               specBg=specBg, 
                               colorOffset=colorOffset, 
                               showErrorOnlyEvery=showErrorOnlyEvery)
        scPlot=OLEDNamedListPlot(prefix,
                                 labelList,
                                 name, 
                                 xLim=xLims[n], 
                                 showColAxLim=yLims, 
                                 showColTup=(n,m), 
                                 spectraFile=spectraPath, 
                                 scaleX=scaleX, 
                                 customFontsize=scaledCustomFontsize, 
                                 titleBool=titleBool, 
                                 show=show, 
                                 errors=errors, 
                                 ax2Labels=ax2Labels, 
                                 alphaOffset=alphaOffset, 
                                 samples=samples, 
                                 specBg=specBg, 
                                 colorOffset=colorOffset, 
                                 showErrorOnlyEvery=showErrorOnlyEvery)
    else:
        plot=OLEDNamedListPlot(prefix,
                               labelList,
                               name, 
                               showColTup=(n,m), 
                               spectraFile=spectraPath, 
                               show=show, 
                               errors=errors, 
                               customFontsize=customFontsize, 
                               alphaOffset=alphaOffset, 
                               samples=samples, 
                               specBg=specBg, 
                               colorOffset=colorOffset, 
                               showErrorOnlyEvery=showErrorOnlyEvery)
        scPlot=OLEDNamedListPlot(prefix,
                                 labelList,
                                 name, 
                                 showColTup=(n,m), 
                                 spectraFile=spectraPath, 
                                 scaleX=scaleX, 
                                 customFontsize=scaledCustomFontsize, 
                                 titleBool=titleBool, 
                                 show=show, 
                                 errors=errors, 
                                 ax2Labels=ax2Labels, 
                                 alphaOffset=alphaOffset, 
                                 samples=samples, 
                                 specBg=specBg, 
                                 colorOffset=colorOffset, 
                                 showErrorOnlyEvery=showErrorOnlyEvery)
    return [plot,scPlot]
```

```python
multiPlot=MultiPlot([initPlot(2, m=3)]+[initPlot(n) for n in range(4,7)])
```

```python
files=multiPlot.getFiles()
```
