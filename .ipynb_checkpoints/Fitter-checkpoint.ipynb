```python
import numpy as np
import scipy.optimize as sco
import copy
from Data import Data
```

```python
class Fitter:
    
    def __init__(self, data, function):
        self.data = data
        self.function = function
        self.dataForFit = copy.deepcopy(data)
        self.params = None
        self.params_cov = None
        self.CurveData = None
        
    def limitInputData(self, xLim=None, yLim=None):
        """deprecated"""
        self.dataForFit.limitData(xLim=xLim, yLim=yLim)
        
    def limitData(self, xLim=None, yLim=None, feature=0):
        if feature == 0:
            self.data.limitData(xLim=xLim, yLim=yLim)
        elif feature == 1:
            self.dataForFit.limitData(xLim=xLim, yLim=yLim)
        elif feature == 2:
            self.CurveData.limitData(xLim=xLim, yLim=yLim)
        else:
            raise ValueError("No Such Feature")
    
    def fit(self):
        self.params, self.params_cov = sco.curve_fit(self.function, *self.dataForFit.getSplitData2D())
    
    def doFitCurveData(self):
        xdata = self.data.getSplitData2D()[0]
        x_min = np.amin(xdata)
        x_max = np.amax(xdata)
        d_x = x_max-x_min
        x_padding_fac = 0.05
        x_padding = x_padding_fac*d_x
        fit_x_data = np.linspace(x_min-x_padding/2, x_max+x_padding/2, num=10**4)
        fit_y_data = self.function(fit_x_data, *self.params)
        self.CurveData = Data.initData2D(fit_x_data, fit_y_data)
```
