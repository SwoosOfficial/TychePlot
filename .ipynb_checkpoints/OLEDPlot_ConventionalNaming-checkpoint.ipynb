```python
#obligatory
xLims=[[],[-2,10],[-2,6],[-2,6],[2.3,6],[2.3,6],[2.3,6]]
yLims=[[],[-2,10],[10**-7,1],[10**-2,10**5],[0,250],[0,600],[0,25]]
spectraPath="/home/tassilo/Dokumente/data/20171204_XXXVIII_JETI_Irppy/XXXVIII_C_100muA.csv"
prefix="XXXVIII"
title="Ir(ppy)$_\\text{3}$"
labelList=[#"Ir(ppy)$_\text{3}$ OLED", 
        #"CsPbBr$_\\text{3}$ QD spin coated at 2000rpm",
        #"CsPbBr$_\\text{3}$ QD spin coated at 3000rpm", 
        "Ir(ppy)$_\\text{3}$ OLED"]
```

```python
plot1=OLEDNamedListPlot(prefix,labelList,title, xLim=xLims[2], yLim=[yLims[2],yLims[3]], showColTup=(2,3), spectraFile=spectraPath, show=[[True,True]]*4, customFontsize=[12,12,8,10,10], alphaOffset=alphaOffset, devices=devices, specBg=5, colorOffset=1)
plot2=OLEDNamedListPlot(prefix,labelList,title,xLim=xLims[2], yLim=[yLims[2],yLims[3]], showColTup=(2,3), spectraFile=spectraPath, scaleX=0.47, customFontsize=[10,10,4,6,6], titleBool=False, show=[[True,True]]*4, ax2Labels=False, alphaOffset=alphaOffset, devices=devices, specBg=5, colorOffset=1)
plot1.doPlot()
plot2.doPlot()
```
