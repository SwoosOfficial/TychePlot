```python
import matplotlib.pyplot as plt
import copy
import sys
sys.path.append("/home/tassilo/Vorlagen/Calculator/")
from matplotlib import rc
from Filereader import fileToNpArray
from Data import Data
from Fitter import Fitter
```

```python
import numpy as np
#import scipy as sci
#import scipy.optimize as sco
```

```python
data, desc = fileToNpArray("/media/tassilo/LUBUNTU 16_/data/Pedot_tissue_tip.csv", separator=",", skiplines=40, backoffset=2, lastlines=2)
data = Data(data, desc=desc)
```

```python
def func(x, a, b):
    return a*x+b
```

```python
offset=[0,0]
fit_xLim1=[7500,8200]
fit_xLim2=[2000,2220]
fit_yLim2=[0,900]
xLim1=[6500,8500]
xLim2=[2000,4000]
```

```python
oriFitter = Fitter(data, func)
fitter = copy.deepcopy(oriFitter)
fitter2 = copy.deepcopy(oriFitter)
fitter.limitInputData(xLim=fit_xLim1)
fitter.limitData(xLim=xLim1)
fitter2.limitInputData(xLim=fit_xLim2, yLim=fit_yLim2)
fitter2.limitData(xLim=xLim2)
fitter.data.offsetData(offset)
fitter.fit()
fitter2.fit()
fitter.doFitCurveData()
fitter2.doFitCurveData()
```

```python
plt.close("all")
f, ax = plt.subplots(2)
plt.rc("text", usetex=True)
plt.rc("font", family="serif")
ax[1].set_xlabel("\\textbf{Length} ($\mu m$)")
ax[0].set_ylabel("\\textbf{Height} (\AA)")
ax[1].set_ylabel("\\textbf{Height} (\AA)")
a, = ax[0].plot(*fitter.data.getSplitData2D(), "r,", label="DEKTAK Topography")
b, = ax[0].plot(*fitter.CurveData.getSplitData2D(), label="Linear Fit")
c, = ax[1].plot(*fitter2.data.getSplitData2D(), "r,", label="DEKTAK Topography")
d, = ax[1].plot(*fitter2.CurveData.getSplitData2D(), label="Linear Fit")
ax[0].legend(handles=[a,b], loc=1)
ax[1].legend(handles=[c,d], loc=1)
plt.show()
```

```python
def modfunc(ydata, xdata=fitter.data.getSplitData2D()[0], params=fitter.params, function=fitter.function):
    return ydata-function(xdata, *params)
```

```python
fitter.data.modifyData(modfunc)
fitter.CurveData.modifyData(modfunc, xdata=fitter.CurveData.getSplitData2D()[0])
```

```python
fitter2.data.modifyData(modfunc, xdata=fitter2.data.getSplitData2D()[0], params=fitter2.params)
fitter2.CurveData.modifyData(modfunc, xdata=fitter2.CurveData.getSplitData2D()[0], params=fitter2.params)
```

```python
plt.close("all")
f, ax = plt.subplots(2)
plt.rc("text", usetex=True)
plt.rc("font", family="serif")
ax[1].set_xlabel("\\textbf{Length} ($\mu m$)")
ax[0].set_ylabel("\\textbf{Height} (\AA)")
ax[1].set_ylabel("\\textbf{Height} (\AA)")
a, = ax[0].plot(*fitter.data.getSplitData2D(), "r,", label="DEKTAK Topography")
b, = ax[0].plot(*fitter.CurveData.getSplitData2D(), label="Linear Fit")
c, = ax[1].plot(*fitter2.data.getSplitData2D(), "r,", label="DEKTAK Topography")
d, = ax[1].plot(*fitter2.CurveData.getSplitData2D(), label="Linear Fit")
ax[0].legend(handles=[a,b], loc=1)
ax[1].legend(handles=[c,d], loc=1)
plt.show()
```

```python
def const(x, a):
    return a+0*x
```

```python
horiFit_xLim1=[7100,7400]
horiFit_yLim1=[250, 750]
horiFit_xLim2=[2400, 2700]
horiFit_yLim2=[250, 750]
```

```python
horiFitter = Fitter(fitter.data, const)
horiFitter2 = Fitter(fitter2.data, const)
horiFitter.limitInputData(xLim=horiFit_xLim1, yLim=horiFit_yLim1)
horiFitter2.limitInputData(xLim=horiFit_xLim2, yLim=horiFit_yLim2)
horiFitter.fit()
horiFitter2.fit()
horiFitter.doFitCurveData()
horiFitter2.doFitCurveData()
```

```python
plt.close("all")
f, ax = plt.subplots(2)
plt.rc("text", usetex=True)
plt.rc("font", family="serif")
ax[1].set_xlabel("\\textbf{Length} ($\mu m$)")
ax[0].set_ylabel("\\textbf{Height} (\AA)")
ax[1].set_ylabel("\\textbf{Height} (\AA)")
a, = ax[0].plot(*fitter.data.getSplitData2D(), "r,", label="DEKTAK Topography")
b, = ax[0].plot(*fitter.CurveData.getSplitData2D(), label="Linear Fit")
b2, = ax[0].plot(*horiFitter.CurveData.getSplitData2D(), "g-", label="Linear Fit")
c, = ax[1].plot(*fitter2.data.getSplitData2D(), "r,", label="DEKTAK Topography")
d, = ax[1].plot(*fitter2.CurveData.getSplitData2D(), label="Linear Fit")
e, = ax[1].plot(*horiFitter2.CurveData.getSplitData2D(), "g-", label="Linear Fit")
ax[0].legend(handles=[a, b, b2], loc=1)
ax[1].legend(handles=[c, d, e], loc=1)
plt.show()
```

```python
delta1 = int(np.round(horiFitter.params/10, decimals=0)[0])
delta1
```

```python
delta2 = int(np.round(horiFitter2.params/10)[0])
delta2
```

```python
fitter.limitData(xLim=[7400, 8400], feature=2)
horiFitter.limitData(xLim=[7000, 7550], feature=2)
fitter2.limitData(xLim=[0,2400], feature=2)
horiFitter2.limitData(xLim=[2200, 2800], feature=2)
```
