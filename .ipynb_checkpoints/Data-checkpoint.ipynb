```python
class Data:
    
    def __init__(self, data, desc=None, xCol=1, yCol=2):
        self.__data = data
        self.xCol = xCol
        self.yCol = yCol
        self.desc = desc

        
    def __iter__(self):
        return getData()
        
    def getData(self): 
        return self.__data
    
    def setData(self, data):
        self.__data = data
        
    
    def getSplitData2D(self):
        data = self.getData()
        xdata=data[range(0,len(data)),[self.xCol-1]*len(data)]
        ydata=data[range(0,len(data)),[self.yCol-1]*len(data)]
        return xdata, ydata
    
    @classmethod
    def initData2D(cls, xdata, ydata):
        a=np.asarray([xdata,ydata], dtype=np.float64)
        return Data(a.transpose())
    
    @classmethod
    def mergeData2D(cls, xdata, ydata):
        a=np.asarray([xdata,ydata], dtype=np.float64)
        return a.transpose()
    
    
    def offsetData(self, offset):
        xdata, ydata = self.getSplitData2D()
        xdata = xdata+offset[0]
        ydata = ydata+offset[1]
        self.setData(Data.mergeData2D(xdata, ydata))
        
    def modifyData(self, function, x=False, y=True, **kwargs):
        xdata, ydata = self.getSplitData2D() 
        if (x == True):
            xdata = function(xdata, **kwargs)
        if (y == True):
            ydata = function(ydata, **kwargs)
        self.setData(Data.mergeData2D(xdata, ydata))
            
    def limitData(self, xLim=None, yLim=None):
        data = self.getData()
        if isinstance(xLim, list):
            if len(xLim) == 2:
                data=data[(data[:,0]>xLim[0]) & (data[:,0]<xLim[1])]
        if isinstance(yLim, list):
            if len(yLim) == 2:
                data=data[(data[:,1]>yLim[0]) & (data[:,1]<yLim[1])]
        self.setData(data)
        
```
