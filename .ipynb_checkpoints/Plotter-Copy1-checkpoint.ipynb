```python
import numpy as np
import scipy as sci
import scipy.optimize as sco
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
def fileToNpArray(filename, separator=",", skiplines=1, backoffset=0, lastlines=0):
    file = open(filename,"r")
    lines = file.readlines()
    lines = lines[:-lastlines]
    file.close()
    yAxis = len(lines)
    preArray = []
    descList = []
    fileToNpArray.array = []
    try:
        #assert(lines[-1].find(separator) != -1)
        for x in range(0, skiplines):
            oriItem = lines[x]
            item = oriItem[:-1]
            if separator != " ":
                item = item.replace(" ","")
            line = item.split(separator)
            if separator == " ":
                line = list(filter(None, line))
            descList.append(line)
        for x in range(skiplines, yAxis):
            oriItem = lines[x]
            realBackoffset=-(1+backoffset)
            item = oriItem[:realBackoffset]
            if separator != " ":
                item = item.replace(" ","")
            line = item.split(separator)
            if separator == " ":
                line = list(filter(None, line))
            preArray.append(line)
        print(preArray[-4:])
        fileToNpArray.array = np.asarray(preArray, dtype=np.float64)
    #except AssertionError:
        #separator = input("Separator \"" + separator + "\" not valid\nPlease specify the right one:")
        #fileToNpArray.array, descList = fileToNpArray(filename, separator=separator, skiplines=skiplines)
    except ValueError:
        skiplinesBoolean = input("Is this the first line of data: " + lines[skiplines] + "\n y/n?")
        if skiplinesBoolean == "y" or skiplinesBoolean == "Y":
            separatorBoolean = input("Is this the separator of data, coulums: \"" + separator + "\"\n y/n?")
            if separatorBoolean == "y" or skiplinesBoolean == "Y":
                raise ValueError("There was an error creating the array from the Data.")
            else:
                separator = input("Please specify the new separator:")
        else:
            skiplinesOperator = input("How many lines are description/nondata? \nType \"next\" for trying the next line.")
            if skiplinesOperator.isdigit():
                skiplines = int(skiplinesOperator)
            else:
                skiplines += 1 
        fileToNpArray.array, descList = fileToNpArray(filename, separator=separator, skiplines=skiplines)
    return fileToNpArray.array, descList
```

```python
data, desc = fileToNpArray("/media/tassilo/LUBUNTU 16_/data/Pedot_tissue_tip.csv", separator=",", skiplines=40, backoffset=2, lastlines=2)
```

```python
xdataLim=[6500,8500]
data=data[(data[:,0]>xdataLim[0]) & (data[:,0]<xdataLim[1])]
```

```python
ydataLim=[0,1]
data=data[(data[:,1]>ydataLim[0]) & (data[:,1]<ydataLim[1])]
```

```python
xCol = 1 
yCol = 2
xdata = data[range(0,len(data)),[xCol-1]*len(data)]
ydata = data[range(0,len(data)),[yCol-1]*len(data)]
```

```python
x_min=np.amin(xdata)
x_max=np.amax(xdata)
y_min=np.amin(ydata)
y_max=np.amax(ydata)
d_x=x_max-x_min
d_y=y_max-y_min
x_padding_fac=0.05
y_padding_fac=0.1
x_padding=x_padding_fac*d_x
y_padding=y_padding_fac*d_y
```

```python
def test_exp(x, a, b):
    return a*np.exp(b*x)
```

```python
fit_x_data = np.linspace(x_min-x_padding/2, x_max+x_padding/2, num=10**4)
```

```python
params, params_cov = sco.curve_fit(test_exp, xdata, ydata)
```

```python
fit_data=test_exp(fit_x_data, *params)
```
