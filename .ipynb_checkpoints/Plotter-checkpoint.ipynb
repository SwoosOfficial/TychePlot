```python
import numpy as np
import scipy as sci
import scipy.optimize as sco
import matplotlib.pyplot as plt
from matplotlib import rc
from Filereader import fileToNpArray
```

```python
data, desc = fileToNpArray("file.dat", separator=" ")
```

```python
xCol = 1 
yCol = 2
xdata = data[range(0,len(data)),[xCol-1]*len(data)]
ydata = data[range(0,len(data)),[yCol-1]*len(data)]
```

```python
x_min=np.amin(xdata)
x_max=np.amax(xdata)
y_min=np.amin(ydata)
y_max=np.amax(ydata)
d_x=x_max-x_min
d_y=y_max-y_min
x_padding_fac=0.05
y_padding_fac=0.1
x_padding=x_padding_fac*d_x
y_padding=y_padding_fac*d_y
```

```python
def test_exp(x, a, b):
    return a*np.exp(b*x)
```

```python
fit_x_data = np.linspace(x_min-x_padding/2, x_max+x_padding/2, num=10**4)
```

```python
params, params_cov = sco.curve_fit(test_exp, xdata, ydata)
```

```python
fit_data=test_exp(fit_x_data, *params)
```

```python
plt.figure(1)
plt.rc("text", usetex=True)
plt.rc("font", family="serif")
plt.xlabel("\\textbf{Time} ($s$)")
plt.ylabel("Eskalation")
a, = plt.plot(xdata, ydata, "ro", label="\\LaTeX", )
b, = plt.plot(fit_x_data, fit_data, label="fit")
plt.legend(handles=[a,b])
table = plt.table(cellText="hallo")
plt.show()
```
